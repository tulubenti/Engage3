# Ridge regression
Ridge regression reduces model complexity and prevent over-fitting which may result from simple linear regression. 
This model shrinks the coefficients and it helps to reduce the model complexity and multi-collinearity. 
In ridge regression, the cost function is altered by adding a penalty equivalent to square of the magnitude of the coefficients.

```python
from sklearn.linear_model import Ridge
```

```python
# Training model
```

```python
ridgeReg = Ridge(alpha=0.0001, normalize=True)
ridgeReg1 = Ridge(alpha = 0.01, normalize = True)
ridgeReg2 = Ridge(alpha = 1, normalize = True)
```

```python
ridgeReg.fit(X_train,y_train)
ridgeReg1.fit(X_train,y_train)
ridgeReg2.fit(X_train,y_train)
```

```python
pred = ridgeReg.predict(X_test)
pred1 = ridgeReg1.predict(X_test)
pred2 = ridgeReg2.predict(X_test)
```

```python
# Calculating mse
```

```python
from sklearn import metrics
```

```python
mse_ridge = metrics.mean_squared_error(y_test,pred)
mse_ridge1 = metrics.mean_squared_error(y_test,pred1)
mse_ridge2 = metrics.mean_squared_error(y_test,pred2)
```

```python
mse_rig =[mse_ridge, mse_ridge1, mse_ridge2]
alpha =['alpha = 0.0001', 'alpha = 0.01', 'alpha = 1']
mse_rig
```

```python
MSE_RIG = pd.Series(mse_rig,alpha).sort_values()
```

```python
MSE_RIG.plot(kind ='bar')
plt.ylim(0.9, 1.05)
```

```python
ridgeReg.coef_
```

```python
cdf_ridge = pd.DataFrame(ridgeReg.coef_, X.columns, columns =['Ridge Coeff'])
```

```python
cdf_ridge
```

```python
# Checking the magnitude of coefficients
```

```python
predictors = X_train.columns
```

```python

ridge_coef = pd.Series(ridgeReg.coef_,predictors).sort_values()
ridge_coef1 = pd.Series(ridgeReg1.coef_,predictors).sort_values()
ridge_coef2 = pd.Series(ridgeReg2.coef_,predictors).sort_values()
```

```python
ridge_coef.plot(kind='bar', title='Alpha = 0.01')
ridge_coef1.plot(kind='bar', title='Alpha = 0.01')
ridge_coef2.plot(kind='bar', title='Alpha = 0.01')
```

# Lasso (Least Absolute Shrinkage and Selection Operator) regression regression
Lasso regression not only helps in reducing over-fitting but it can help us in feature selection.

```python
from sklearn.linear_model import Lasso
```

```python
lassoReg = Lasso(alpha= 0.0001, normalize=True)
lassoReg1 = Lasso(alpha= 0.01, normalize=True)
lassoReg2 = Lasso(alpha= 1, normalize=True)
```

```python
lassoReg.fit(X_train,y_train)
lassoReg1.fit(X_train,y_train)
lassoReg2.fit(X_train,y_train)
```

```python
predLasso = lassoReg.predict(X_test)
predLasso1 = lassoReg1.predict(X_test)
predLasso2 = lassoReg2.predict(X_test)
```

```python
mse_lasso = metrics.mean_squared_error(y_test,predLasso)
mse_lasso1 = metrics.mean_squared_error(y_test,predLasso1)
mse_lasso2 = metrics.mean_squared_error(y_test,predLasso2)
```

```python
MSE_LASSO = [mse_lasso,mse_lasso1, mse_lasso2]
alphaLasso =['alpha = 0.0001', 'alpha = 0.01', 'alpha = 1']
MSE_LASSO
```

```python
MSE_LASS = pd.Series(MSE_LASSO,alpha).sort_values()
```

```python
MSE_LASS.plot(kind ='bar')
# plt.ylim(0.9, 1.05)
```
